```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from minisom import MiniSom
from sklearn.preprocessing import MinMaxScaler
```

```python
dataset = pd.read_csv("../Self_Organizing_Maps/Credit_Card_Applications.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
sc = MinMaxScaler(feature_range=(0,1))
X = sc.fit_transform(X)
```

```python
#input_len => number of independent variable
#sigma is the radius of the different radius in the grid
#learning_rate => how much the weight updated for each iteration
som = MiniSom(x = 15,
              y = 15,
              input_len=15,
              sigma=1.0, 
              learning_rate=0.5)

#we need to initialize the weight
som.random_weights_init(X)
som.train_random(data = X, num_iteration=100)
```

```python
from pylab import bone, pcolor, colorbar, plot, show
bone()
#som.distance_map returns all the distance of each winning node
#each winning node is represented as square in graph
pcolor(som.distance_map().T)
#show the mean inter neuron distances
colorbar()
'''
    we can define the MID equals to 1 as the outlier of winning nodes
        why?
        because we can explain that the frauds are identified by
        outlaying winning nodes those are far from general rules
        our dark color is close each other which means our general
        rules is dark or smallest MID
        
    the smallest mean corresponds to dark color
    the highest mean corresponds to light color
    
'''
#o = red circles as doesn't get approval,
#s = green square as customer get approval
#fraud is the customer get approved but cheated
markers = ['o', 's']
colors = ['r', 'g']

#iterate through array using enumerate
for i, x in enumerate(X) :
    winning_node = som.winner(x)
    plot(winning_node[0] + 0.5, 
         winning_node[1] + 0.5, 
         markers[y[i]], 
         markeredgecolor = colors[y[i]],
         markerfacecolor = 'None',
         markersize = 10,
         markeredgewidth = 2)
show()
```

```python
'''
    mappings is dictionary where the key represents the coordinate of
    winning nodes
    the key contains list of all member of that nodes
    e.g. (5,9) : List of 28 members 
                where each member represents row in the X
'''
mappings = som.win_map(X)
```

```python
frauds = np.concatenate((mappings[(1,2)], mappings[(3,8)]), axis=0)
```

```python
frauds = sc.inverse_transform(frauds)
'''this customer can be identified as fraudulent customer since they got approved 
but they are different with majorities'''
frauds
```

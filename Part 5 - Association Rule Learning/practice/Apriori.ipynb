```python
from apyori import apriori
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Section 28 - Apriori/Market_Basket_Optimisation.csv", header=None)
df.tail()
```

```python
transactions = []
for i in range(0, 7501) :
    transactions.append([str(df.values[i, j]) for j in range(0,20)])
```

```python
'''
if product is purchased 3 times per week which means
    min_support = 3 * 7 / total_transactions
    min_support = 21 / 7500
    min_support = 0.003
    
    confidence formula => cart contains (p1, p2)/total cart contains p1
    confidence should not too high since we will eliminate all the rules that might be exist
    min_confidence = 20%
    
    min_lift =3
    
    length is the product combination
'''

#apriory function import
rules = apriori(transactions, \
               min_support= 0.003,\
               min_confidence= 0.15,\
               min_lift=2,\
               max_length=3)
```

```python
result = list(rules)
result[0:5]
```
